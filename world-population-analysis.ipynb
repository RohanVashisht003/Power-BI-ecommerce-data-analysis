```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("countries-table.csv")
```

```python
print(df)
```

```python
# Remove null values
df=df.dropna()
```

```python
df.head()
```

```python
# Plot 1: Top 10 most populated countries
top_10_countries=df.nlargest(10,columns="pop2023")
plt.figure(figsize=(12,6))
sns.barplot(x="country",y="pop2023",data=top_10_countries)
plt.title("Top 10 countries by population 2023")
plt.xlabel("Country")
plt.ylabel("Population(in billions)")
plt.xticks(rotation=90)
plt.show()
```

```python
# Plot 2: Top 10 most populated countries trend year wise
countries = top_10_countries["country"]
years= ["pop1980","pop2000","pop2010","pop2022","pop2023","pop2030","pop2050"]

plt.figure(figsize=(15,15))

for country in countries:
    country_data=df[df["country"] == country]
    plt.plot(years,country_data[years].values.reshape(-1),label=country)
      
plt.xlabel("Years")
plt.ylabel("Population (in billion)")
plt.title("Population trend for top 10 highly populated countries")
plt.legend()
plt.show()
```

```python
top_10_population_growth_rate = df.sort_values('growthRate', ascending=False).head(10)
plt.bar(top_10_population_growth_rate["country"],top_10_population_growth_rate["growthRate"],color="orange")
plt.xlabel("Country")
plt.ylabel("Growth rate")
plt.title("Top 10 countries with highest population increase in 2023")
plt.xticks(rotation=45)
plt.show()
```
